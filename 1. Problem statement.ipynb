### Problem Statement
Artificial Intelligence (AI) technologies are rapidly transforming industries worldwide. Automation tools, machine learning systems, and intelligent algorithms are increasingly capable of performing tasks that were previously done by humans. Because of this transformation:
- Some jobs may disappear
- Some jobs may change significantly
- Some jobs may increase in demand

However, organizations, workers, and policymakers often do not clearly know which jobs are at higher risk of automation. This creates several challenges including

- Workers may not know which skills to learn
- Companies may not know which roles will change
- Governments may not know which industries need reskilling programs

### Purpose of the Project

The goal of this project is to use Data Mining and Machine Learning techniques to analyze job market data and predict the automation risk of occupations. The project aim to
- Analyze patterns in job-related data
- Identify factors that influence automation risk
- Build predictive models that classify job roles based on their risk level

### Project Objectives

This project aim to predicting which jobs are likely to be replaced or affected by AI and machines in the future by using Data Mining and Machine Learning Techniques

### Problem Statement
Artificial Intelligence (AI) is growing very quickly and changing how many industries work. Because of this, some jobs may be replaced or heavily affected by automation process of AI. However, it is difficult to clearly identify which jobs are more likely at higher risk.

This project aims to apply data mining techniques to analyze job market data and develop a predictive model that identifies the level of automation risk for different occupations.

### The main questions this project tries to answer are:

- Can machine learning models predict the automation risk category of jobs?
- Which factors (such as AI adoption level, industry type, and skill gap index) most influence job automation risk?
- How do different industries and job roles vary in their vulnerability to AI automation?

# Predicting Job Automation Risk Using Machine Learning

### Problem Statement

AI is changing many jobs. Some may disappear, some will change, and others may grow.

- Workers don’t know which skills to learn.

- Companies don’t know which roles will change.

- Governments don’t know which industries need reskilling programs.

Goal: Use machine learning to predict the automation risk of jobs and identify key factors affecting it.

### Data Collection

Dataset: Future of Work in the Age of AI (2020–2026)

Dataset link:
https://www.kaggle.com/datasets/algozee/future-of-work-in-the-age-of-ai-20202026

- Columns include: job_role, industry, country, automation_risk_percent, skill_gap_index, salary_before_usd, salary_after_usd, ai_adoption_level, automation_risk_category, etc.

### Data Preprocessing

- Feature creation: salary_change = salary_after_usd - salary_before_usd

- Feature engineering: Convert text to numbers (job_role, industry, country, ai_adoption_level, automation_risk_category)

- Feature selection: Drop columns closely related to the target to prevent overfitting.

- Train-test split: 80% training, 20% testing.

### Overall Analysis

- Decision Tree achieved the highest accuracy (97.2%), indicating it captures patterns in the training data very well.

- Random Forest, despite being a robust ensemble method, showed imbalanced precision/recall across classes, resulting in lower overall accuracy (84.3%).

- The results suggest feature engineering, class distribution, and hyperparameters play a key role in Random Forest performance.

- For deployment or reporting, Decision Tree is the best performing model, but Logistic Regression can be considered for more interpretable results.

After evaluating multiple models including Logistic Regression, Decision Tree, and Random Forest we selected Logistic Regression for training because

- Interpretable, allowing clear understanding of how each feature influences automation risk.

- Achieves good performance (94.4% accuracy) without overfitting, unlike tree-based models with very high training accuracy.

- Efficient and scalable, making it suitable for deployment on larger datasets.

Overall, Logistic Regression effectively predicts job automation risk while keeping the model transparent and actionable, helping workers, companies, and policymakers plan for AI-driven changes.

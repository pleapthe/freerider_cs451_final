```python
import pandas as pd

# Load Excel dataset
df = pd.read_excel("cam-key-data.xlsx")

# Display first rows
print(df.head())

# Convert wide format to long format
df_long = df.melt(
    id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
    var_name="Year",
    value_name="Value"
)

# Convert Year to numeric
df_long["Year"] = pd.to_numeric(df_long["Year"], errors="coerce")

# Sort values
df_long = df_long.sort_values(["Indicator Name", "Year"])

# Display cleaned data
print(df_long.head())

# Save cleaned dataset
df_long.to_csv("cambodia_cleaned_data.csv", index=False)
```

```python
import pandas as pd

# -------------------------------
# 1. Load Dataset
# -------------------------------
df = pd.read_csv("cambodia_cleaned_data.csv")
print("Initial dataset:")
print(df.head())
print(df.info())

# -------------------------------
# 2. Inspect Missing Values
# -------------------------------
print("\nMissing values per column:")
print(df.isnull().sum())

# -------------------------------
# 3. Handle Missing Values
# -------------------------------

# Convert 'Year' to int
df["Year"] = pd.to_numeric(df["Year"], errors="coerce").astype(int)

# Ensure 'Value' is numeric
df["Value"] = pd.to_numeric(df["Value"], errors="coerce")

# Sort by Indicator and Year
df = df.sort_values(["Indicator Name", "Year"])

# Interpolate missing values per indicator
df["Value"] = df.groupby("Indicator Name")["Value"].transform(lambda x: x.interpolate(method='linear'))

# Optional: Drop any remaining missing values
df = df.dropna(subset=["Value"])

# -------------------------------
# 4. Remove Duplicates
# -------------------------------
df = df.drop_duplicates()

# -------------------------------
# 5. Optional: Split by Indicator Type
# -------------------------------
economic_indicators = ["GDP", "Inflation", "Unemployment", "Trade"]
social_indicators = ["Population", "Life expectancy", "Education"]

df_economic = df[df["Indicator Name"].str.contains("|".join(economic_indicators), case=False, na=False)]
df_social = df[df["Indicator Name"].str.contains("|".join(social_indicators), case=False, na=False)]

# -------------------------------
# 6. Save Cleaned Datasets
# -------------------------------
df.to_csv("cambodia_cleaned_final.csv", index=False)
df_economic.to_csv("cambodia_economic_indicators.csv", index=False)
df_social.to_csv("cambodia_social_indicators.csv", index=False)

# -------------------------------
# 7. Display cleaned dataset info
# -------------------------------
print("\nCleaned dataset preview:")
print(df.head(10))
print(df.info())
print("\nMissing values after cleaning:")
print(df.isnull().sum())
```
